```python
plt.figure(figsize=(8,5))
subset = coffee_df.groupby("Year")["Sales"].sum().reset_index()

# Wrong variable for title, truncated y-axis, misleading color & framing
plt.plot(subset["Year"], subset["Sales"], color="red", linewidth=4, marker="o")
plt.ylim(subset["Sales"].min() * 0.98, subset["Sales"].min() * 1.02)  # extreme truncation
plt.title("Shocking Collapse in Profits!", fontsize=16, fontweight="bold", color="darkred")
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10,6))
avg_sales = coffee_df.groupby("Region")["Sales"].mean().reset_index()



plt.bar(avg_sales["Region"], avg_sales["Sales"], 
        color="#05a763", edgecolor="#035230", linewidth=3, hatch="**", alpha=0.7)
plt.title("Average Sales by Region", fontsize=20, fontweight="bold", pad=15, color="#04003b")
plt.xlabel("Region", fontsize=16, fontweight="bold", color="#0C0333")
plt.ylabel("Average Sales", fontsize=16, color="#0C0333")
plt.xticks(rotation=0, fontsize=12, fontweight="bold", color="black")
plt.yticks(fontsize=12,  color="black")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
avg_profit = coffee_df.groupby("Region")["Profit"].mean().reset_index()



plt.bar(avg_profit["Region"], avg_profit["Profit"], 
        color="#b256e7", edgecolor="#7F0EC0", linewidth=3, hatch="**", alpha=0.7)
plt.title("Average Profit by Region", fontsize=20, fontweight="bold", pad=15, color="#04003b")
plt.xlabel("Region", fontsize=16, fontweight="bold", color="#03330F")
plt.ylabel("Average Profit", fontsize=16, color="#03330F")
plt.xticks(rotation=0, fontsize=12, fontweight="bold", color="black")
plt.yticks(fontsize=12,  color="black")

plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(8,5))
subset = coffee_df.groupby("Year")["Sales"].sum().reset_index()

# Wrong variable for title, truncated y-axis, misleading color & framing
plt.plot(subset["Year"], subset["Sales"], color="#5E0606", linewidth=3, marker="o")
plt.ylim(subset["Sales"].min(), subset["Sales"].min())  # extreme truncation
plt.title("Sales from 2020 to 2024 \n Is this dip in Sales caused by Covid?", pad= 25, fontsize=15, fontweight="bold", color="#032902", backgroundcolor="#f3baaf")
plt.xlabel("Year", fontsize=14, fontweight="bold", color= "#311702")
plt.ylabel("Sales", fontsize=14, fontweight="bold", color= "#2c2903")
plt.xticks(fontsize=10, color= "#290205")
plt.yticks(fontsize=10, color= "#290205")
ax= plt.gca()
for spine in ax.spines.values():
    spine.set_linewidth(1.5)
    spine.set_alpha(0.5)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)


plt.tight_layout()

plt.show()
```

Could be other causes, but if there were no major issues with the business in that time, it would appear Covid is the culprit.
